# Scraping

```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
tickers = ['TCS.NS', 'HDFCBANK.NS', 'RELIANCE.NS', 'SUNPHARMA.NS', 'ITC.NS']
print("Downloading data")
data = yf.download(tickers, start = "2014-01-01", end = "2023-12-31")['Close']
data.head()
```

```python
print(f"Number of missing values: {data.isna().sum().sum()}")

returns = data.dropna()
returns.head()
```

```python
print("Annualised Expected Returns:")
mean_returns = returns.mean() * 252
print(mean_returns)
```

```python
print("Annualised Risk: ")
std_returns = returns.std() * np.sqrt(252)
print(std_returns)
```

```python
cov_matrix = returns.cov() * 252
print("Annualised Covariance Matrix\n")
print(cov_matrix)
```

```python
# Save data to CSV 
returns.to_csv('stock_data.csv')
mean_returns.to_csv('mean_returns.csv')
std_returns.to_csv('std_returns.csv')
cov_matrix.to_csv('cov_matrix.csv')
```
