# Scraping

```python
import yfinance as yf
import pandas
import numpy
```

```python
tickers = ['NIFTYBEES.NS', 'MID150BEES.NS', 'MON100.NS', 'GOLDBEES.NS', 'LIQUIDBEES.NS']
print("Downloading data")
data = yf.download(tickers, period = '5y')['Close']
data.head()
```

```python
returns = data.pct_change().dropna()
returns.head()
```

```python
mean_returns = returns.mean() * 252
print("Annualised Mean Returns\n")
print(mean_returns)
```

```python
cov_matrix = returns.cov() * 252
print("Annualised Covariance Matrix\n")
print(cov_matrix)
```

```python
data.to_csv("historical_prices.csv")
mean_returns.to_frame(name='Mean Returns').to_csv('mean_returns.csv')
cov_matrix.to_csv('cov_matrix.csv')
print("Data successfully saved to CSV files!")
```
