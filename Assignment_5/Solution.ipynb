```python
goal_a = {
    "goal_1": (1500000, 3),
    "goal_2": (2500000, 7), 
    "goal_3": (3000000, 12),
    "terminal": (15000000, 20)
}
goal_b = {
    "goal_1": (1000000, 8),
    "goal_2": (2000000, 12), 
    "goal_3": (4000000, 16),
    "terminal": (15000000, 20)
}
```

```python
discrete_weights = [0.0, 0.25, 0.5, 0.75, 1.0]
```

# Task 2.1 - All possible weights

```python
all_combinations = []
for w1 in discrete_weights:
    for w2 in discrete_weights:
        for w3 in discrete_weights:
            for w4 in discrete_weights:
                for w5 in discrete_weights:
                    if w1 + w2 + w3 + w4 + w5 == 1.0:
                        all_combinations.append((w1, w2, w3, w4, w5))

print(f"Total valid combinations: {len(all_combinations)}")
print("Possible Combinations:")
for combo in all_combinations:
    print(combo)
```

# Task 2.2 - Monte Carlo Simulation

```python
def run_monte_carlo(weights, mean_returns, cov_matrix, goals, num_paths=5000, years=20):
    # 1. Calculate Portfolio Expected Return and Standard Deviation
    # Using matrix multiplication for the specific weight combination
    port_return = np.dot(weights, mean_returns)
    port_std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    
    # 2. Generate Random Returns for all paths and all years at once (Vectorized)
    # Shape will be (5000 paths, 20 years)
    np.random.seed(42) # Keeps results consistent for grading
    simulated_returns = np.random.normal(port_return, port_std, (num_paths, years))
    
    # 3. Track portfolio balances across all 5000 paths
    balances = np.zeros(num_paths)
    
    # Run the simulation year by year
    for year in range(1, years + 1):
        # A. Add Annual Savings (₹20,000/month annualized, growing at 4%)
        annual_savings = initial_savings_per_month * 12 * ((1 + savings_increase) ** (year - 1))
        balances += annual_savings
        
        # B. Apply Market Growth or Borrowing Penalty
        returns_this_year = simulated_returns[:, year - 1]
        
        # Identify which paths are in the green vs. which are in debt
        is_positive = balances > 0
        
        # Positive balances grow by the simulated market return
        balances[is_positive] = balances[is_positive] * (1 + returns_this_year[is_positive])
        
        # Negative balances (debt) grow by the strict 12% shortfall interest rate
        balances[~is_positive] = balances[~is_positive] * (1 + shortfall_interest_rate)
        
        # C. Deduct Goal Amounts if due in the current year
        for goal_name, (amount, goal_year) in goals.items():
            if year == goal_year:
                balances -= amount
                
    # 4. Calculate Probability of Success
    # Success is defined as having a final balance >= 0 AFTER the terminal goal is deducted
    successful_paths = np.sum(balances >= 0)
    probability_of_success = successful_paths / num_paths
    
    return probability_of_success
```

```python
# Create an empty list to store our results
results_list = []

# Flatten the mean returns so it is a 1D array (5,) instead of a 2D matrix (5, 1)
mean_ret_array = mean_returns.values.flatten()
cov_mat_array = cov_matrix.values

for combo in all_combinations:
    # You can directly convert the combo tuple into a numpy array
    weights = np.array(combo)
    
    # Run the simulations
    prob_a = run_monte_carlo(weights, mean_ret_array, cov_mat_array, goal_a)
    prob_b = run_monte_carlo(weights, mean_ret_array, cov_mat_array, goal_b)
    
    # Append the results as a dictionary to our list
    results_list.append({
        'Weights': weights, 
        'Prob_Success_Goal_A': prob_a, 
        'Prob_Success_Goal_B': prob_b
    })

# Convert the populated list into a DataFrame all at once
newdf = pd.DataFrame(results_list)

# Display the final results
print(newdf)

# Optional: To easily spot your best portfolios, sort the DataFrame!
print("\nTop 5 Portfolios for Sequence A:")
print(newdf.sort_values(by='Prob_Success_Goal_A', ascending=False).head())

print("\nTop 5 Portfolios for Sequence B:")
print(newdf.sort_values(by='Prob_Success_Goal_B', ascending=False).head())
```
