```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
tickers = [
    "TCS.NS", "INFY.NS", "HCLTECH.NS", "TATASTEEL.NS", "HINDALCO.NS", "JINDALSTEL.NS", "HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS", "HINDUNILVR.NS", "ITC.NS", "NESTLEIND.NS"
]
```

```python
data = yf.download(
    tickers,
    period="5y",        
    interval="1d",      
    auto_adjust=False,  
    group_by="ticker",
    threads=True
)


prices = pd.DataFrame()

for ticker in tickers:
    prices[ticker] = data[ticker]["Adj Close"]
prices = prices.dropna()

daily_log_returns = np.log(prices/prices.shift(1))
daily_log_returns = daily_log_returns.dropna()
```

```python
monthly_prices = prices.resample("M").last()
monthly_log_returns = np.log(monthly_prices / monthly_prices.shift(1))
monthly_log_returns = monthly_log_returns.dropna()
```

```python
last_date = daily_log_returns.index.max()
split_date = last_date - pd.DateOffset(years=1)
```

```python
price_train = prices[prices.index <= split_date]
price_test  = prices[prices.index > split_date]

daily_train = daily_log_returns[daily_log_returns.index <= split_date]
daily_test  = daily_log_returns[daily_log_returns.index > split_date]

monthly_train = monthly_log_returns[monthly_log_returns.index <= split_date]
monthly_test  = monthly_log_returns[monthly_log_returns.index > split_date]

price_train.to_csv("../data/price_train.csv")
price_test.to_csv("../data/price_test.csv")

daily_train.to_csv("../data/daily_train.csv")
daily_test.to_csv("../data/daily_test.csv")

monthly_train.to_csv("../data/monthly_train.csv")
monthly_test.to_csv("../data/monthly_test.csv")
```
